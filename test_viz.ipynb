```python
from image_processor import ImageProcessor
import os
import cv2 as cv
```

```python
def load_videos(path: str) -> tuple:

    screen_frames = [f'{path}/screen/{frame}' 
                        for frame in sorted(os.listdir(f'{path}/screen'))
                            if frame != '.DS_Store'][94:100]
    
    hm_frames = [f'{path}/heatmap/{frame}' 
                    for frame in sorted(os.listdir(f'{path}/heatmap')) 
                        if frame != '.DS_Store'][94:100]

    sal_frames = [f'{path}/sal/{frame}' 
                    for frame in sorted(os.listdir(f'{path}/sal')) 
                        if frame != '.DS_Store'][94:100]

    return sorted(screen_frames), sorted(hm_frames), sorted(sal_frames)
```

```python
screen_frames, hm_frames, sal_frames = load_videos('../data/demo/E4')
```

```python
screen_frames = list(map(lambda x: cv.imread(x), screen_frames))
hm_frames = list(map(lambda x: cv.imread(x), hm_frames))
sal_frames = list(map(lambda x: cv.imread(x, cv.IMREAD_GRAYSCALE), sal_frames))
```

```python
image_processor = ImageProcessor(screen_frames, hm_frames, sal_frames)
```
